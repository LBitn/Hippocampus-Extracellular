```julia
A = reverse( reshape( collect(1:4096), 64, 64 )', dims = 1 );
using StatsBase
x = sample(1:4096, 300)
```

```julia
orillaizq = collect(1:64:4033)
orillaarr = collect(4033:4096)
orillader = collect(64:64:4096)
orillaaba = collect(1:64)
grupos = []
estado = false
for i in 1:64
    for j in 1:64
        grupovac = []
        estado = false
        if A[i,j] in x
        if A[i,j] in orillaizq && A[i,j] in orillaaba
            for grupo in grupos
                for elemento in grupo    
                    if A[i-1,j] == elemento || A[i-1,j+1] == elemento || A[i,j+1] == elemento
                        push!(grupo,A[i,j])
                        estado = true
                    end
                end
            end
            if estado == false
                push!(grupovac,A[i,j])
                push!(grupos,grupovac)
            end
        elseif A[i,j] in orillaizq && A[i,j] in orillaarr
            for grupo in grupos
                for elemento in grupo    
                    if A[i,j+1] in grupo || A[i+1,j-1] in grupo || A[i+1,j] in grupo
                        push!(grupo,A[i,j])
                        estado = true
                    end
                end
            end
            if estado == false
                push!(grupovac,A[i,j])
                push!(grupos,grupovac)
            end
        elseif A[i,j] in orillader && A[i,j] in orillaarr
            for grupo in grupos
                for elemento in grupo    
                    if A[i,j-1] in grupo || A[i+1,j-1] in grupo || A[i+1,j] in grupo
                        push!(grupo,A[i,j])
                        estado = true
                    end
                end
            end
            if estado == false
                push!(grupovac,A[i,j])
                push!(grupos,grupovac)
            end
        elseif A[i,j] in orillader && A[i,j] in orillaba
            for grupo in grupos
                for elemento in grupo    
                    if A[i,j-1] in grupo || A[i-1,j-1] in grupo || A[i-1,j] in grupo
                        push!(grupo,A[i,j])
                        estado = true
                    end
                end
            end
            if estado == false
                push!(grupovac,A[i,j])
                push!(grupos,grupovac)
            end   
        elseif A[i,j] in orillaizq
             for grupo in grupos
                for elemento in grupo    
                    if A[i-1,j] in grupo || A[i-1,j+1] in grupo || A[i,j+1] in grupo || A[i+1,j+1] in grupo || A[i+1,j] in grupo
                        push!(grupo,A[i,j])
                        estado = true
                    end
                end
            end
            if estado == false
                push!(grupovac,A[i,j])
                push!(grupos,grupovac)
            end
        elseif A[i,j] in orillaarr
            for grupo in grupos
                for elemento in grupo    
                    if A[i,j-1] in grupo || A[i+1,j-1] in grupo || A[i+1,j] in grupo || A[i+1,j+1] in grupo || A[i,j+1] in grupo
                        push!(grupo,A[i,j])
                        estado = true
                    end
                end
            end
            if estado == false
                push!(grupovac,A[i,j])
                push!(grupos,grupovac)
            end    
        elseif A[i,j] in orillader
            for grupo in grupos
                for elemento in grupo    
                    if A[i-1,j] in grupo || A[i-1,j-1] in grupo || A[i,j-1] in grupo || A[i+1,j-1] in grupo || A[i+1,j] in grupo
                        push!(grupo,A[i,j])
                        estado = true
                    end
                end
            end
            if estado == false
                push!(grupovac,A[i,j])
                push!(grupos,grupovac)
            end     
        elseif A[i,j] in orillaaba
            for grupo in grupos
                for elemento in grupo    
                    if A[i,j-1] in grupo || A[i-1,j-1] in grupo || A[i-1,j] in grupo || A[i-1,j+1] in grupo || A[i,j+1] in grupo
                        push!(grupo,A[i,j])
                        estado = true
                    end
                end
            end
            if estado == false
                push!(grupovac,A[i,j])
                push!(grupos,grupovac)
            end    
        else
            for grupo in grupos
                for elemento in grupo    
                    if A[i-1,j-1] in grupo || A[i-1,j] in grupo || A[i-1,j+1] in grupo || A[i,j-1] in grupo || A[i,j+1] in grupo || A[i+1,j-1] in grupo || A[i+1,j] in grupo || A[i+1,j+1] in grupo
                        push!(grupo,A[i,j])
                        estado = true
                    end
                end
            end
            if estado == false
                push!(grupovac,A[i,j])
                push!(grupos,grupovac)
            end    
        end
        end
    end
end
println(grupos)
```
